```python
df3= df2.drop(['Cabin','Name','PassengerId','Ticket'],axis=1)
```

```python
df.shape , df2.shape , df3.shape 
```

```python
df3['Embarked'].value_counts()
```

```python
df3['Sex'].value_counts()
```

```python
df3['Embarked'] = df3['Embarked'].replace('S',1)
df3['Embarked'] = df3['Embarked'].replace('C',2)
df3['Embarked'] = df3['Embarked'].replace('Q',3)
df3['Sex'] = df3['Sex'].replace('male',1)
df3['Sex'] = df3['Sex'].replace('female',2)
```

```python
df3['Sex'].value_counts()
```

```python
df3['Embarked'].value_counts()
```

```python
df3.head()
```

```python
x = df3.drop('Survived',axis=1)
y = df3['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.model_selection import train_test_split
```

```python
x.shape , y.shape
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.20, random_state=45)
```

```python
model=[LogisticRegression(), DecisionTreeClassifier(criterion='gini'), DecisionTreeClassifier(criterion='entropy'),
      SVC(kernel='rbf'), GaussianNB()]
```

```python
for i in range(5,15):
    print('\n')
    print('n_neighbors= ', i)
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_pred=knn.predict(x_test)
    print('Accuracy Score: ', accuracy_score(y_pred,y_test))
    print('\n')
    print(confusion_matrix(y_pred,y_test))
    print('\n')
    print('Classification Report: ', classification_report(y_pred,y_test))
    print('\n')
    print('\n')
```

```python
## max accuracy at n=13 which is 73%
```

```python
for m in model:
    print('\n')
    print('Model= ', m)
    m.fit(x_train, y_train)
    y2=m.predict(x_test)
    print('Accuracy Score: ', accuracy_score(y2,y_test))
    print('\n')
    print(confusion_matrix(y2,y_test))
    print('\n')
    print('Classification Report: ', classification_report(y2,y_test))
    print('\n')
    print('\n')
```

```python
## gaussianNB is giving best results i.e 0.8048
```

```python
print('cv_folds= ', 5)
for m in model:
        print('\n')
        print('Model= ', m)
        print('\n')
        score=cross_val_score(m,x,y,cv=i)
        print('Score= ', score)
        print('\n')
        print(score.mean(), score.std())
        print('\n')
        y_pred=cross_val_predict(m,x,y,cv=i)
        print('Accuracy Score: ', accuracy_score(y,y_pred))
        print('\n')
        print(confusion_matrix(y,y_pred))
        print('\n')
        print('Classification Report: ', classification_report(y,y_pred))
        print('\n')
        print('\n')
```

```python
cross_val_knn = cross_val_score(knn,x,y,cv=5).mean()
cross_val_knn
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('cross_val_rfc' , cross_val_score(rfc,x,y,cv=5).mean())
```

```python
params = {'n_estimators' : np.arange(100,120,2),
          'criterion' : ['gini', 'entropy'],
          'min_samples_split' : np.arange(1,10),
          'min_samples_leaf' : np.arange(1,5),
          'max_features' : ['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
GCV = GridSearchCV(rfc,params,n_jobs=-1)
GCV.fit(x_train,y_train)
GCV.best_params_
```

```python
rfc_model = RandomForestClassifier(criterion='entropy',max_features='sqrt',min_samples_leaf=2,min_samples_split=5,n_estimators=116)
rfc_model.fit(x_train,y_train)
```

```python
y_pred_rfc = rfc_model.predict(x_test)
print(accuracy_score(y_pred_rfc,y_test))
print(confusion_matrix(y_pred_rfc,y_test))
print(classification_report(y_pred_rfc,y_test))
```

```python
print('rfc_cross_val_score' , cross_val_score(rfc_model,x,y,cv=5).mean())
```
